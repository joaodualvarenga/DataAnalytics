Prática de Redes Neurais

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn import datasets
import numpy as np
```

```python
Diagnostico = pd.read_csv('soybean.csv')
Diagnostico.head()
Diagnostico.shape
```

```python
previsores = Diagnostico.iloc[:,0:35].values
classe = Diagnostico.iloc[:,35].values
```

```python
labelencoder = LabelEncoder()
for x in range(35):
    previsores[:,x] = labelencoder.fit_transform(previsores[:,x])
previsores = np.asarray(previsores).astype(np.float32)
previsores
```

```python
labelencoder_Y = LabelEncoder()
classe = labelencoder_Y.fit_transform(classe)
classe = np.asarray(classe).astype(np.float32)
classe
```

```python
classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

```python
unicos, quantidade = np.unique(classe, return_counts = True)
unicos
```

```python
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores,
                                                                  classe_dummy,
                                                                  test_size = 0.3,
                                                                  random_state = 0)
```

```python
# Criação da estrutura da rede neural com a classe Sequential (sequência de camadas)
modelo = Sequential()
#primeira camada oculta, 5 neuronios, 4 neuronios de entrada
modelo.add(Dense(units = 36, input_dim = 35))
#segunda camada oculta
modelo.add(Dense(units = 42))
modelo.add(Dense(units = 27))
# Função softmax porque temos um problema de classificação com mais de duas classes 
#(é gerada uma probabilidade em cada neurônio)
modelo.add(Dense(units =19 , activation = 'softmax'))
```

```python
modelo.summary()
```

```python
modelo.compile(optimizer = 'adam', loss = 'categorical_crossentropy',
               metrics = ['accuracy'])
# Treinamento, dividindo a base de treinamento em uma porção para validação (validation_data)
modelo.fit(X_treinamento, y_treinamento, epochs = 1000,
           validation_data = (X_teste, y_teste))
```

```python
previsoes = modelo.predict(X_teste)
previsoes = (previsoes > 0.5)
previsoes
```

```python
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
```

```python
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
confusao
```

```python
taxa_acerto = accuracy_score(y_teste, previsoes)
taxa_acerto
```
