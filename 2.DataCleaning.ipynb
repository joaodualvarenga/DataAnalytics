```python
mediana1 = est.median(tabela['Umidade'])
mediana1
```

```python
tabela['Umidade'].fillna(mediana1,inplace=True)
```

```python
desv1 = est.stdev(tabela['Umidade'])
desv1
```

```python
sea.boxplot(tabela['Umidade'])
```
